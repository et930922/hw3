```python
import pandas as pd
df=pd.read_csv("datasets/algebradata.csv")
df.head()
```

```python
# Question A: %age of students with a passing grade (Answer: 91.6917%)
```

```python
def score_to_numeric(x):
    if x=='A':
        return 1
    if x=='B':
        return 1
    if x=='C':
        return 1
    if x=='D':
        return 1
    if x=='F':
        return 2
```

```python
df['Grade_R']=df['Grade'].apply(score_to_numeric)
df.head()
```

```python
A=pd.crosstab(index=df["Grade_R"],columns="count") 
A
```

```python
A/A.sum()
```
