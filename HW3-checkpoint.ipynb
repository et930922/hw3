```python
import pandas as pd
df=pd.read_excel('titanic3.xlsx')
df.head()
```

```python
df.describe()
```

```python
#Fare variable-the max is 512 which seems an outlier so remove that record(s) from analysis
```

```python
newdf=df
newdf=newdf.drop(newdf[newdf['fare']==512.329200].index)
newdf.describe()
```

```python
# There're 4 records which fare is 512.329200 so we removed 4 records and created new datset "newdf"
```

```python
newdf.corr()
```

```python
# Use regerssion to know what drives people survived. 
# First, try to put pclass,age,sibsp,parch,fare,and sex as independent variables and survived as dependent variable
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ pclass +age +sibsp +parch +fare +sexR', data=newdf).fit()
result.summary()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ pclass +age +sibsp +parch +fare +sexR-1', data=newdf).fit()
result.summary()
```

```python
# 1) Based on correlation results table,it seems pclass, fare, parch and sex shows relatively higher correlation with survived,
# 2) Based on regression, parch and fare are excluded due to P value is higher than 5% which is not significant.
# 3) Remove parch & fare from regression model and see results again
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ pclass +age +sibsp +sexR', data=newdf).fit()
result.summary()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ pclass +age +sibsp +sexR-1', data=newdf).fit()
result.summary()
```

```python
# 2nd regression model shows all the four independ variables(pclass, age, sib & sex) P value less than 0.05 (significant). 
# Thus, we use desectiptive statistics to know survived people's pclass, age, sib and sex information
```

```python
# Finding A: More females (72%) survived than males (18%).
A= pd.crosstab(index=newdf["survived"], columns=newdf["sex"])
A.index= ["died","survived"]
A/A.sum()
```

```python
# Finding B: Highier classes more likely to survived 
B= pd.crosstab(index=newdf["survived"], columns=newdf["pclass"])
B.index= ["died","survived"]
B/B.sum()
```

```python
# Finding C: More sibsp is more likely to died 
C= pd.crosstab(index=newdf["survived"], columns=newdf["sibsp"])
C.index= ["died","survived"]
C/C.sum()
```

```python
# Finding D_1: Average suvived people age is 28.8 yrs old
survived_age=newdf[newdf['survived']==1]
survived_age.describe()
```

```python
# Finding D_2: Average died people age is 30.5 yrs old which is slightly higher than survived group
died_age=newdf[newdf['survived']==0]
died_age.describe()
```

```python
# Finding F: Based on corsstable below, 96% of 1st class female survived which is highest group of people who survived
# More males in 2nd and 3rd class were died than 1st class make. 
F= pd.crosstab(index=newdf["survived"], columns=[newdf["pclass"],newdf["sex"]])
F/F.sum()
```
